```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import guided_project_functions as gpf
```

```python
# Read the data
hotel_bookings = pd.read_csv("hotel_bookings.csv")
```

```python
# Explore the data
hotel_bookings.head()
```

```python
#Describe the data
hotel_bookings.describe()
```

```python
# Print info about the data
hotel_bookings.info()
```

```python
#Find the nulls values
hotel_bookings.isnull().sum().sort_values(ascending=False)
```

```python
#  Dropping the company and the agent columns, seems unnecessary because they have a lot of nulls.
hotel_bookings.drop(hotel_bookings[['company', 'agent']], axis=1, inplace=True)
```

```python
# Check the changes.
hotel_bookings.shape
```

```python
# Count the values in the country column.
hotel_bookings['country'].value_counts()
```

```python
# Replace the nulls with 'PRT', because this is the common country.
hotel_bookings['country'] =hotel_bookings['country'].fillna('PRT')
```

```python
# Check if there are still nulls.
gpf.count_nulls('country')
```

```python
# Replace the nulls in the children column to 0 because there are only 4 nulls.
hotel_bookings['children'] =hotel_bookings['children'].fillna(0)
```

```python
# Check if there are still nulls.
hotel_bookings['children'].isnull().sum()
```

```python
# Change the month's name to numbers
hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date_month'].map({'January' : 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November': 11, 'December':12})
```

```python
# Create a column of arrival date, including day, month and year.
hotel_bookings['arrival_date'] =pd.to_datetime(dict(year=hotel_bookings.arrival_date_year, month=hotel_bookings.arrival_date_month, day=hotel_bookings.arrival_date_day_of_month))
```

```python
#Find the correlation between lead time and is canceles.
sns.heatmap(data=hotel_bookings[['lead_time','is_canceled']].corr(),annot=True, cmap='Purples')
plt.suptitle("Lead time Impacting Cancellations")
```

From the heatmap, we see that there is almost no correlation between the lead time and the bookings that were canceled.

```python
# Create a data frame with the bookings that weren't canceled.
hotel_bookings_not_canceled = hotel_bookings[hotel_bookings['is_canceled']==0]
hotel_bookings_not_canceled.shape
```

```python
# Print the columns names.
hotel_bookings.columns
```

```python
# Create a catplot to find the booking per month for each hotel.
sns.catplot(data = hotel_bookings_not_canceled, x = 'arrival_date_month', col= 'hotel', kind= 'count',palette='rainbow')
plt.suptitle("Booking Per Month", y=1.05)
```

We see that generally there are more bookings in the City Hotel. In the two hotels, the highest season is August, and September and January are the lowest seasons, in the Resort Hotel also June, September and November are low.

```python
# Create a catplot to find the booking per year for each hotel.
sns.catplot(data = hotel_bookings_not_canceled, x = 'arrival_date_year', col='hotel', kind= 'count',palette='rainbow')
plt.suptitle("Booking Per Year")
```

We can see that both hotels had the most bookings in 2016.

```python
# A function to make a column of locals and of people from other places.
def local(x):
    if x == 'PRT':
        return 'local'
    else:
        return 'Out of town'
```

```python
# Apply the function.
hotel_bookings_not_canceled['local'] = hotel_bookings_not_canceled['country'].apply(local)
```

```python
# Create a catplot to find how many people in each hotel are locals and how many are not.
sns.catplot(data=hotel_bookings_not_canceled, x='local', kind='count', hue='hotel')
```

We found that most of the people aren't locals. And that the locals go to the 2 hotels almost equally.

```python
# Count how many are from each country.
hotel_bookings_not_canceled['country'].value_counts().head(10)
```

Although in general there are more non-locals than locals, when we look at each country by itself, we see that Portugal is in first place, which means that relatively there are the most locals.
