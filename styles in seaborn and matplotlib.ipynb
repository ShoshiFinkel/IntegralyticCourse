```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
donuts = pd.read_csv('donuts.csv')
```

```python
donuts.head()
```

```python
# In the 'bmh' style there are grids aligning to the ticks.
style.use('bmh')
donuts['branch'].value_counts().plot.bar(color = 'purple')  
plt.suptitle('Donuts Branches', weight='bold')
plt.xlabel("Branch")
plt.ylabel("Donuts Sold")
plt.xticks(rotation=0)
```

```python
#In the classic style there is a white background to the bars and outside the bar it's colored, which makes the bar very clear.
style.use('classic')
donuts['branch'].value_counts().plot.bar(color = 'purple')  
plt.suptitle('Donuts Branches', weight='bold')
plt.xlabel("Branch")
plt.ylabel("Donuts Sold")
plt.xticks(rotation=0)
```

```python
style.use('default')
```

```python
sns.catplot(data=donuts, x='branch', kind='count', hue='type')
plt.suptitle('Donuts Filling Per Branch',weight='bold')
sns.set_style('ticks')
sns.set_palette('muted')
sns.set_context('poster')
```

```python
style.use('default')
```

```python
# The ticks style made the text bigger and easy for read. The pastel pallete made the colors pastely
sns.catplot(data=donuts, x='branch', kind='count', hue='type')
plt.suptitle('Donuts Filling Per Branch',weight='bold',y=1.05)
sns.set_style('white')
sns.set_palette('pastel')
sns.set_context('paper')
```
