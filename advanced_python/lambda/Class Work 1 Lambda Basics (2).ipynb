#### Section A
Convert these def functions to lambda functions

```python
#1 Convert to lambda function
def add_15(num):
    total = num+15
    return total

l_add_15 = lambda num: num + 15
```

```python
#2
#Call the lambda on any odd number.
l_add_15(3)
```

```python
#3 Convert to lambda function
def calculate_cube_root(x):
    return x**(1/3)

l_calculate_cube_root=lambda x: x**(1/3)
```

```python
#4
#call the lambda on the number 64 and round the answer to two decimal places.
round(l_calculate_cube_root(64), 2)
```
