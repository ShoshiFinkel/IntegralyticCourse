```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data
hotel_bookings = pd.read_csv("hotel_bookings.csv")
```

```python
# Explore the data
hotel_bookings.head()
```

```python
hotel_bookings.describe()
```

```python
hotel_bookings.info()
```

```python
#Find the nulls values
hotel_bookings.isnull().sum().sort_values(ascending=False)
```

```python
# Dropping the company and the agent columns, they seems unrelavent becouse they have a lot of nulls.
hotel_bookings.drop(hotel_bookings[['company', 'agent']], axis=1, inplace=True)
```

```python
# Check the changes
hotel_bookings.shape
```

```python
hotel_bookings['country'] =hotel_bookings['country'].fillna('Unknown')
```

```python
hotel_bookings['country'].isnull().sum()
```

```python
hotel_bookings['children'] =hotel_bookings['children'].fillna(0)
```

```python
hotel_bookings['children'].isnull().sum()
```

```python
hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date_month'].map({'January' : 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November': 11, 'December':12})
```

```python
hotel_bookings['arrival_date'] =pd.to_datetime(dict(year=hotel_bookings.arrival_date_year, month=hotel_bookings.arrival_date_month, day=hotel_bookings.arrival_date_day_of_month))
```

```python
sns.heatmap(data=hotel_bookings[['lead_time','is_canceled']].corr(),annot=True, cmap='Purples')
plt.suptitle("Lead time Impacting Cancellations")
```

```python
hotel_bookings.columns
```

```python
sns.catplot(data = hotel_bookings, x = 'arrival_date_month', kind= 'count',palette='rainbow')
plt.suptitle("Booking Per Month")
```

```python
sns.catplot(data = hotel_bookings, x = 'arrival_date_year', kind= 'count',palette='rainbow')
plt.suptitle("Booking Per Year")
```

```python
country = hotel_bookings.groupby('country')
```

```python
sns.catplot(data=country,x=country)
```

```python
def local(x):
    if x == 'PRT':
        return 'local'
    else:
        return 'Out of town'
```

```python
hotel_bookings['local'] = hotel_bookings['country'].apply(local)
```

```python
sns.relplot(data=hotel_bookings,x="hotel", y="bill_length_mm",hue = 'local', style = "sex",size = 'flipper_length_mm', sizes= (15, 200))
plt.suptitle("Booking Per Year")
```
