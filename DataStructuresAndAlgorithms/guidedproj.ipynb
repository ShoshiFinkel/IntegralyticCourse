```python
import csv
import time
open_file= open('fake_job_postings.csv', encoding = 'utf-8')
read_file=csv.reader(open_file)
fake_job_postings=list(read_file)
print(fake_job_postings[0])
```

```python
industry = []
for row in fake_job_postings:
 #   if row[-3] != 'Farming' and row[-3] != 'Mining & Metals' and row[-3] != 'Entertainment':
    industry.append(row[-3])
for each in industry:
    if each == 'Farming' or each == 'Mining & Metals' or each == 'Entertainment':
        industry.remove(each)
industry.extend(['Sofer', 'Teacher', 'Secratery', 'Cosmetics', 'Music'])

set_industry = set(industry)

def count_industry_list(a_set, a_list):
    for each in a_set:
        if each != '':
            print(each, a_list.count(each))

start = time.time()
count_industry_list(set_industry, industry)
end = time.time()
list_time = end - start
print(list_time)
```

```python
def count_industry_dict(a_set, a_list):
    count_dict = {}
    for each in a_set:
        if each != '':
            count_dict[each] = a_list.count(each)
    return count_dict

start = time.time()
count_industry_dict(set_industry, industry)
end = time.time()
dict_time = end - start
print(dict_time)
```

```python
title = []
for row in fake_job_postings:
    title.append(row[1])
tuple_title = tuple(title)
set_tuple_title = set(tuple_title)

def count_title_tuple(a_set, a_tuple):
    for each in a_set:
        if each != '':
            print(each, a_tuple.count(each))

start = time.time()
count_title_tuple(set_tuple_title, title)
end = time.time()
tuple_time = end - start
print(tuple_time)      
    
```

```python
location = []
for row in fake_job_postings:
    location.append(row[2])
location = sorted(location)


def binary_search(L, target):
    start = 0
    end = len(L) - 1

    while start <= end:
        middle = (start + end)// 2
        midpoint = L[middle]
        if midpoint > target:
            end = middle - 1
        elif midpoint < target:
            start = middle + 1
        else:
            return True
    return False



target = 'AE, DU, Dubai'
target = 'ISRAEL'
start = time.time()
result = binary_search(location, target)
print(result)
end = time.time()
l_binary_search = end - start
print(l_binary_search)
```

```python
location_set = set(location)
location_set = sorted(location_set)


def set_binary_search(S, target):
    start = 0
    end = len(S) - 1

    while start <= end:
        middle = (start + end)// 2
        midpoint = S[middle]
        if midpoint > target:
            end = middle - 1
        elif midpoint < target:
            start = middle + 1
        else:
            return True
    return False



target = 'AE, DU, Dubai'
#target = 'ISRAEL'
start = time.time()
result = set_binary_search(location_set, target)
print(result)
end = time.time()
s_binary_search = end - start
print(s_binary_search)
```
