```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
donuts = pd.read_csv('donuts.csv')
```

```python
donuts.head()
```

```python
# In the 'bmh' style there are grids aligning to the ticks.
style.use('bmh')
donuts['branch'].value_counts().plot.bar(color = 'purple')  
plt.suptitle('Donuts Branches', weight='bold')
plt.xlabel("Branch")
plt.ylabel("Donuts Sold")
plt.xticks(rotation=0)
```

```python
#In the classic style there is a white background to the bars and outside the bar it's colored, which makes the bar very clear.
style.use('classic')
donuts['branch'].value_counts().plot.bar(color = 'purple')  
plt.suptitle('Donuts Branches', weight='bold')
plt.xlabel("Branch")
plt.ylabel("Donuts Sold")
plt.xticks(rotation=0)
```

```python
# The whitegrid style shows grids on a white background which makes it easy to read. The muted palette made the colors dark with a white line. and it has big text so it is easy to read.
sns.catplot(data=donuts, x='branch', kind='count', hue='type')
plt.suptitle('Donuts Filling Per Branch',weight='bold')
sns.set_style('whitegrid')
sns.set_palette('muted')
sns.set_context('poster')
```

```python
# The darkgrid style shows white grids on a dark background which makes it easy to read. The pastel palette made the colors pastel with a white line.
sns.catplot(data=donuts, x='branch', kind='count', hue='type')
plt.suptitle('Donuts Filling Per Branch',weight='bold',y=1.05)
sns.set_style('darkgrid')
sns.set_palette('pastel')
sns.set_context('paper')
```
